```python
def plot(original: dict, processed: dict, output:str, save:bool=False):
    
    fname = subject['fname']
    fig, ax = plt.subplots(2, 1, sharex=False, figsize=(16,8), tight_layout=True)
    t1 = original['onsets'].flatten() - original['onsets'].flatten()[0]
    ax[0].set_title(f"Conditions: subject {fname}")
    # ax[0].plot(t1, original["trials"].flatten() / 8, label="trials")
    ax[0].plot(t1, original["sleep"].flatten(), label="sleep")
    ax[0].plot(t1, original["fussy"].flatten(), label="fussy")
    ax[0].plot(t1, original["volume"].flatten(), label="volume")
    ax[0].set_xticks([t1[0], t1[-1]])
    ax[0].set_xticklabels([f"{t1[0]:.3f}", f"{t1[-1]:.3f}"])
    ax[0].set_ylabel("Value")
    ax[0].set_xlabel("Time (s)")
    ax[0].legend()

    t2 = processed['onsets'].flatten() - processed['onsets'].flatten()[0]
    # ax[1].plot(t2, processed["trials"].flatten() / 8, label="trials")
    ax[1].plot(t2, processed["sleep"].flatten(), label="sleep")
    ax[1].plot(t2, processed["fussy"].flatten(), label="fussy")
    ax[1].plot(t2, processed["volume"].flatten(), label="volume")
    ax[1].set_xticks([t2[0], t2[-1]])
    ax[1].set_xticklabels([f"{t2[0]:.3f}", f"{t2[-1]:.3f}"])
    ax[1].set_ylabel("Value")
    ax[1].set_xlabel("Time (s)")
    fig.set_size_inches(12, 6)
    ax[1].legend()
  
    if save:
        output /= 'inspection/' 
        output.mkdir(exist_ok=True, parents=True)
        plt.savefig(output / fname / '.png', dpi=300, bbox_inches="tight",)
    plt.tight_layout()
    plt.show(block=False)
    plt.pause(0.1)
  
```

```python
# SETTING UP CLASSES INSTANCIATION
config = Config(root_dir='../')
eeg = EEG(n_channels=config.N_CHANNELS, sfreq=config.S_FREQ)

# DATA READING
# CLEAN DATA
root = os.path.abspath('./')
basename = "/" + os.path.basename(root) 
config.ROOT_PATH = root

subject_files = eeg.get_subject_files(
        data_path=Path(str(config.CLEAN_SUBJECTS_PATH).replace(basename, '')) / "train", ext=config.FILE_EXT
    )
```

```python
for filename in tqdm(subject_files, colour="BLUE", desc="Inspection...:"):
      subject = eeg.get_subject(
          filename=filename, time_points=config.TIME_POINTS, sfreq=config.S_FREQ
      )
      config.TIMES = subject["t"].flatten()


      processed = eeg.drop_trials(
      subject=subject,
      drop_trials=False,
      include=None,
      drop_volume=True,
      drop_fussy=True
      )
      
      plot(subject, processed, Path(str(config.OUTPUT_DIR).replace(basename,'')), save=False)
  
```
